```python
import gmaps
import pandas as pd
from config import gkey

# Configure gmaps
gmaps.configure(api_key=gkey)
```

```python
# Create aiport dataframe
csvfile = pd.read_csv("../Resources/Airport_Output.csv")

airport_df = pd.DataFrame(csvfile)
del airport_df["Unnamed: 0"]
airport_df.dropna(inplace=True)
airport_df.head()
```

```python
# Store latitude and longitude in locations
locations = airport_df[["Lat","Lng"]]

# Filla NaN values and convert to float
rating = airport_df["Airport Rating"].astype(float)
airport_df
```

```python
# Plot Heatmap

figure_layout = {
    'width': '400px',
    'height': '400px',
    'border': '1px solid black',
    'padding': '1px'
}
fig = gmaps.figure(layout=figure_layout)

fig = gmaps.figure(map_type='HYBRID')
heatmap_layer = gmaps.heatmap_layer(locations, weights=rating, dissipating=False,
    max_intensity=5, point_radius=1)
fig.add_layer(heatmap_layer)
fig
```
