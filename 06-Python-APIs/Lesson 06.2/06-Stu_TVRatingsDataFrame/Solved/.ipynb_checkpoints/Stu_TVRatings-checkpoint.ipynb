```python
#Dependencies
import requests
import json
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

base_url = "https://api.tvmaze.com/search/shows?"
```

```python
#list of tv show titles to query
tv_shows = ["Altered Carbon", "Grey's Anatomy", "This is Us", "The Flash", "Vikings", "Shameless", "Arrow", "Peaky Blinders", "Dirk Gently"]
rating = []

# make iterative requests to TVmaze search endpoint
for show in tv_shows:
    response = requests.get(base_url + "q=" + show).json()
    rating.append(response[0]["show"]["rating"]["average"])
```

```python
# create dataframe
rating_dict = {
    "show" : tv_shows,
    "rating" : rating
}

show_data = pd.DataFrame(rating_dict).set_index("show")
```

```python
# use pandas to create a bar chart from the dataframe
show_data.plot(kind="bar",legend=False)
plt.xlabel("Shows")
plt.ylabel("Average Rating")
plt.title("Average TV Show Ratings")
plt.xticks(rotation=90)
plt.ylim([0,10])
plt.show()
```
