You are given a dataset that contains historical data from purchases of an online store made by 200 customers. In this activity you will put in action your data preprocessing superpowers, also you'll add some new skills needed to start finding customers clusters.

```python
# Initial imports
import pandas as pd
from pathlib import Path
file_path = Path("../Resources/shopping_data.csv")

df = pd.read_csv(file_path)
df
```

```python
df2 = pd.get_dummies(df)
df2 = df2.drop(["Previous Shopper_No", "CustomerID"], axis='columns')
df2 = df2.rename(columns={"Previous Shopper_Yes": "Previous Shopper"})
df2
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
X = df2[["Age", "Annual Income", "Spending Score (1-100)"]]
scaler = StandardScaler().fit(X)
X_scaled = scaler.transform(X)
```

```python
df_scaled = pd.DataFrame(X_scaled, columns = ['Age','Annual Income','Spending Score (1-100)'])
```

```python
df_scaled["Previous Shopper"] = df2["Previous Shopper"]
df_scaled
```

```python
file_path = Path("../Resources/scaled_data.csv")
df_scaled.to_csv(file_path, index=False)
```
