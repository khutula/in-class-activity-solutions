```python
from sklearn.preprocessing import normalize
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt
from pathlib import Path
```

```python
import pandas as pd
file = Path('../Resources/wholesale_customers.csv')
```

```python
df = pd.read_csv(file)
```

```python
df.sample(10)
```

```python
normalized = normalize(df)
df_norm = pd.DataFrame(normalized)
df_norm.columns = df.columns
df_norm
```

```python
mergings = linkage(normalized, method="ward")
```

```python
plt.figure(figsize=(20,8))
dendrogram(mergings, leaf_rotation=90, leaf_font_size=5)
plt.show()
```

```python
model = AgglomerativeClustering(n_clusters=2, affinity="euclidean", linkage="ward")
model.fit_predict(df_norm)
cluster_labels = model.labels_
```

```python
X = df_norm["Milk"]
y = df_norm["Grocery"]

plt.scatter(X, y, c=cluster_labels)
plt.show()
```
