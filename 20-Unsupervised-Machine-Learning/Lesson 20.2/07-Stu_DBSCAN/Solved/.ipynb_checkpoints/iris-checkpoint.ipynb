```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
# Load iris dataset
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
# The iris data
data = iris.data
data[:5]
```

```python
# Feature names
iris.feature_names
```

```python
# Label (iris species) of each sample
iris.target[:5]
```

```python
# Plot sepal length by petal length
plt.scatter(data[:,0], data[:,2])
plt.show()
```

```python
# Color samples by label
plt.scatter(data[:,0], data[:,2], c=iris.target)
plt.show()
```

```python
# Scale the data
scaler = StandardScaler().fit(data)
data_scaled = scaler.transform(data)
```

```python
# Instantiate and train the model
dbscan = DBSCAN(eps=0.65, min_samples=5)
dbscan.fit(data_scaled)
```

```python
# DBSCAN cluster labels
labels = dbscan.labels_
set(labels)
```

```python
df = pd.DataFrame()
df["X"] = data_scaled[:,0]
df["y"] = data_scaled[:,2]
```

```python
# Plot the results, clustered by DBSCAN labels
plt.scatter(df["X"], df["y"], c=labels)
```
