```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
months = np.arange(1, 13)
temp_f = [39, 42, 51, 62, 72, 82, 86, 84, 77, 65, 55, 44]
temp_c = [(temp-32)*0.56 for temp in temp_f]
```

```python
plt.plot(months, temp_f)
plt.xlabel("Months of 2020")
plt.ylabel("Temperature in Farenheit")
plt.show()
```

```python
plt.plot(months, temp_c)
plt.xlabel("Months of 2020")
plt.ylabel("Temperature in Celsius")
plt.show()
```

```python
plt.plot(months, temp_f)
plt.plot(months, temp_c)
plt.xlabel("Months of 2020")
plt.ylabel("Temperature in Celsius")
plt.legend(["Farenheit","Celsius"])
plt.show()
```
