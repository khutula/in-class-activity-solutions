```python
import matplotlib.pyplot as plt
import numpy as np
# Include this line to make plots interactive
%matplotlib inline
```

```python
months = np.arange(1, 13)
temp_f = [39, 42, 51, 62, 72, 82, 86, 84, 77, 65, 55, 44]
temp_c = [(temp-32)*0.56 for temp in temp_f]
```

```python
plt.plot(months, temp_f, label="Farenheit", color="blue", marker="+")
plt.xlabel("Months of 2020")
plt.ylabel("Temperature in Farenheit")
plt.show()
```

```python
plt.plot(months, temp_c, label="Celsius", color="red", marker="s")
plt.xlabel("Months of 2020")
plt.ylabel("Temperature in Celsius")
plt.show()
```

```python
plt.plot(months, temp_f, label="Farenheit", color="blue", marker="+")
plt.plot(months, temp_c, label="Celsius", color="red", marker="s")
plt.xlabel("Months of 2020")
plt.ylabel("Temperature in Celsius")
#plt.legend(["Farenheit","Celsius"])
plt.show()
```
