```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib notebook
```

```python
runtime = np.arange(0, 130, 10)
danger_drop = [9, 8, 90, 85, 80, 70, 70, 65, 55, 60, 70, 65, 50]
rail_gun = [75, 70, 60, 65, 60, 45, 55, 50, 40, 40, 35, 35, 30]
```

```python
plt.plot(runtime, danger_drop, label="Danger Drop", color="red")
plt.plot(runtime, rail_gun, label="RailGun", color="green")
plt.xlabel("Coaster Runtime")
plt.ylabel("Speed (MPH)")
plt.title("Coaster Speed Over Time")
plt.xlim([0,120])
plt.ylim([5,95])
plt.grid()
plt.show()
```
