```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Load in csv
avg_rain_df = pd.read_csv("../Resources/avg_rain_state.csv")
avg_rain_df.head()
```

```python
plt.bar(avg_rain_df["State"], avg_rain_df["Inches"])
plt.show()
```

```python
plt.xticks(rotation=90)
plt.xlabel("State")
plt.ylabel("Rain (Inches)")
plt.title("Average Rainfall")
```
