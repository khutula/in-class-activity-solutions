```python
# Import Dependencies
import numpy as np
import matplotlib.pyplot as plt
```

```python
# DATA SET 1
gyms = ["Crunch", "Planet Fitness", "NY Sports Club", "Rickie's Gym"]
members = [49, 92, 84, 53]
```

```python
plt.bar(gyms,members)
plt.xlabel("Gym")
plt.ylabel("Number of Members")
plt.title("Members by Gym")
plt.ylim([0,100])
plt.show()
```

```python
# DATA SET 2
x_lim = 2 * np.pi
x_axis = np.arange(0, x_lim, 0.1)
sin = np.sin(x_axis)
```

```python
plt.xlabel("Real Numbers from 0 to 2$\pi$")
plt.ylabel("Value of Sine")
plt.title("Sine from 0 to 2$\pi$")
plt.hlines(0,0,x_lim)
plt.xlim=([0,x_lim])
plt.plot(x_axis, sin)
plt.show()
```

```python
# DATA SET 3
gyms = ["Crunch", "Planet Fitness", "NY Sports Club", "Rickie's Gym"]
members = [49, 92, 84, 53]
x_axis = np.arange(0, len(gyms))
colors = ["yellowgreen", "red", "lightcoral", "lightskyblue"]
explode = (0, 0.05, 0, 0)
```

```python
plt.title("Gym Membership Percentage")
plt.axis("equal")
plt.pie(members, explode=explode, labels=gyms, colors=colors, autopct="%1.1f%%")
plt.show()
```

```python
# DATA SET 4
x_axis = np.arange(0, 10, 0.1)
times = []
for x in x_axis:
    times.append(x * x + np.random.randint(0, np.ceil(max(x_axis))))
```

```python
plt.ylabel("Time to Sort (s)")
plt.xlabel("Length of Input Array")
plt.title("Running Time of FakeSort for Sample Input Sizes")

plt.scatter(x_axis, times)
plt.show()
```
