```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

%matplotlib notebook
```

```python
# Read CSV
battle_kings = pd.read_csv("Resources/got.csv")
battle_kings
```

```python
# Get attacker and defender data
attacker_kings = battle_kings["attacker_king"].value_counts()
defender_kings = battle_kings["defender_king"].value_counts()
```

```python
# Get total battle data
total_battles = attacker_kings.add(defender_kings, fill_value=0)
```

```python
# Configure plot and ticks
```

```python
# Set textual properties
```

```python
count_chart = total_battles.plot(kind="bar", color="red", rot=45)
count_chart.set_xlabel("King")
count_chart.set_ylabel("Number of Battles")
count_chart.set_title("Game of Thrones: Battles per King")

plt.show()
plt.tight_layout()
```
