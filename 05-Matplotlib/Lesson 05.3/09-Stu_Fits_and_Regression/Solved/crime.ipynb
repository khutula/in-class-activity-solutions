```python
# Dependencies
from matplotlib import pyplot as plt
from scipy import stats
import numpy as np
import pandas as pd
```

```python
# Load crime data set into pandas
crime_data = pd.read_csv("../Resources/crime_data.csv")
crime_data.head()
```

```python
# Generate a scatter plot of violent crime rate versus year
x_axis = crime_data["year"]
y_axis = crime_data["violent crime rate"]
(slope_violent, intercept_violent, rvalue, pvalue, stderr) = stats.linregress(x_axis, y_axis)
regress_values = x_axis * slope_violent + intercept_violent
line_eq = "y = " + str(round(slope_violent,2)) + "x + " + str(round(intercept_violent,2))

plt.scatter(x_axis,y_axis)
plt.plot(x_axis, regress_values, "r-")
plt.annotate(line_eq,(2000,600),fontsize=15,color="red")
plt.xticks(x_axis, rotation=90)
plt.xlabel("Year")
plt.ylabel("Violent Crime Rate")
plt.title("Violent Crime Rate Over Time")
plt.show()
```

```python
# Generate a facet plot of all 3 figures
fig, (ax1, ax2, ax3) = plt.subplots(3,1,sharex=True)

x_axis = crime_data["year"]
y_axis = crime_data["violent crime rate"]
(slope_violent, intercept_violent, rvalue, pvalue, stderr) = stats.linregress(x_axis, y_axis)
regress_values = x_axis * slope_violent + intercept_violent
line_eq = "y = " + str(round(slope_violent,2)) + "x + " + str(round(intercept_violent,2))
ax1.scatter(x_axis,y_axis)
ax1.plot(x_axis, regress_values, "r-")
ax1.annotate(line_eq,(2000,600),fontsize=15,color="red")
ax1.set_ylabel("Violent Crime Rate")
ax1.set_title("Crime Rates Over Time")

x_axis = crime_data["year"]
y_axis = crime_data[" murder and non-negligent manslaughter rate"]
(slope_murder, intercept_murder, rvalue, pvalue, stderr) = stats.linregress(x_axis, y_axis)
regress_values = x_axis * slope_murder + intercept_murder
line_eq = "y = " + str(round(slope_murder,2)) + "x + " + str(round(intercept_murder,2))
ax2.scatter(x_axis,y_axis)
ax2.plot(x_axis, regress_values, "r-")
ax2.annotate(line_eq,(2000,7),fontsize=15,color="red")
ax2.set_ylabel("Murder Rate")

x_axis = crime_data["year"]
y_axis = crime_data["aggravated assault rate"]
(slope_agg_assault, intercept_agg_assault, rvalue, pvalue, stderr) = stats.linregress(x_axis, y_axis)
regress_values = x_axis * slope_agg_assault + intercept_agg_assault
line_eq = "y = " + str(round(slope_agg_assault,2)) + "x + " + str(round(intercept_agg_assault,2))
ax3.scatter(x_axis,y_axis)
ax3.plot(x_axis, regress_values, "r-")
ax3.annotate(line_eq,(2000,375),fontsize=15,color="red")
ax3.set_ylabel("Agg Assault Rate")
ax3.set_xlabel("Year")
plt.xticks(x_axis,rotation=90)
plt.show()
```

```python
# Calculate the crime rates for 2019
year = 2019
violent_crime_predict = year * slope_violent + intercept_violent
murder_predict = year * slope_murder + intercept_murder
agg_assualt_predict = year * slope_agg_assault + intercept_agg_assault
```
